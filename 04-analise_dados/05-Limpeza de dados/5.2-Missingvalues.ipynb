```python
datas = pd.date_range(start='20190101', periods=60, freq='D')   
df = pd.DataFrame(np.random.randn(60,5), index=datas, columns=list('ABCDE'))
df.head(3)
```

```python
df['F'] = df.A[ df.A > 0 ]  # Assign values from column A to column F where A > 0
df.head(3)
```
