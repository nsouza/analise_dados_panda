```python
import matplotlib.pyplot as plt

# Gera o histograma e captura os valores
counts, bins, patches = plt.hist(d['Age'], bins=10)

# Rtulos dos eixos
plt.xlabel('Idade')
plt.ylabel('Frequncia observada')

# Adiciona rtulo acima de cada barra
for count, patch in zip(counts, patches):
    # Pega o centro da barra
    x = patch.get_x() + patch.get_width() / 2
    y = patch.get_height()
    plt.text(x, y, int(count), ha='center', va='bottom', fontsize=9)

plt.show()
```
