```python
import pandas as pd 
import numpy as np 
```

```python
datas = pd.date_range('20180101', periods=600, freq='D')
df = pd.DataFrame(np.random.randn(600,5), index = datas, columns = list('ABCDe'))
```

```python
df.head(2)
```

```python
#Obtendo os dados apenas da coluna D
df['D'].head(5)
```

```python
#Obter linhas específicas
df.iloc[1:5]
```

```python
# Obter a coluna BCDE
df.loc[:, 'B':'E'].head(5)
```

```python
# Slice retornando um periodo de data e colunas especificas
df.loc['2018-01-10':'2018-01-20', ['B', 'D'] ].head(5)
```

```python
#selecionar apenas uma linha
df.iloc[5]
```

```python
#Pegar valor de determinada linha e coluna  
df.iloc[2:4, 1:3]
```

```python
#Extrair conjunto de linhas
df.iloc[[1,3,5], [0,3]]
```

```python

#Selecionar dados especifico  de uma linha e todas as colunas
df.iloc[1:3, :]
```

```python
#filtar os valores positivos de uma coluna
df[df.A > 0]
```

```python
#selcionar somente dados positivos de todas as colunas
df[df > 0].head(5)
```
